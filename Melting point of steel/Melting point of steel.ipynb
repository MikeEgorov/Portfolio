### Анализ данных.

```python
# Функция для получения данных о датасетах.

def data_analysis(data):
    print(data.info())
    print('-' * 30)
    display(data.head())
    print('-' * 30)
    print(f'Количество строк и столбцов в датасете: {data.shape}')
    print('-' * 30)
    print(f'Пропуски в датасете:\n{data.isna().sum()}')        
    print('-' * 30)
    print(f'Информация о типах данных в датасете:\n{data.dtypes}')
    print('-' * 30)
    print('Распределение численных значений признаков в датасете:')
    display(data.describe().T)
    print('-' * 30)
    print(f'Количество явных дубликатов в датасете: {data.duplicated().sum()}')
    print('<-->' * 15,'\n')          
```

#### Генерация новых признаков.

```python
# Добавим новый признак со значениями полной мощности для каждого этапа нагрева и каждого временного промежутка.

data_arc['Полная мощность'] = (data_arc['Активная мощность'] ** 2 + data_arc['Реактивная мощность'] ** 2) ** 0.5
```

```python
# Рассчитаем в секундах продолжительность каждого этапа нагрева.

data_arc['Время нагрева (сек.)'] = (data_arc['Конец нагрева дугой'] - data_arc['Начало нагрева дугой']).dt.seconds
```

```python
# На основании значений полной мощности и времени (в сек.) добавим новый признак 'Полная работа' для каждого этапа нагрева.

data_arc['full_work'] = data_arc['Полная мощность'] * data_arc['Время нагрева (сек.)']
```

```python
data_arc.head()
```

```python
# Удалим признаки, на основе которых производились расчеты.

data_arc = data_arc.drop(columns=['Начало нагрева дугой', 
                                  'Конец нагрева дугой', 
                                  'Активная мощность', 
                                  'Реактивная мощность', 
                                  'Полная мощность', 
                                  'Время нагрева (сек.)'], axis=1)
```

```python
# Сагрегируем значения полной работы по партиям, просуммировав их.
# Получим значение полной работы, затраченной на соответствующую партию сырья.

data_arc = data_arc.groupby('key').agg({'full_work': 'sum'})
data_arc.head()
```

```python
# Создадим список номеров партий, которые имеют 2 и более значений температуры, чтобы исключить неполные данные с одним значением.

keys = data_temp_new.key.value_counts() > 1
keys = keys[keys].index.tolist()
```

```python
# Оставим в датасете данные только по тем партиям, которые имеют 2 и более значений температуры.

data_temp_new = data_temp_new[data_temp_new.key.isin(keys)]
```

```python
# Сагрегируем по партиям (key) значения температур и времени замера по принципу "первый-последний".

data_temp_new = data_temp_new.groupby(by='key').agg(start_time=('Время замера', 'first'),
                                                    end_time=('Время замера', 'last'),
                                                    start_temp=('Температура', 'first'),
                                                    end_temp=('Температура', 'last')
                                                   )
```

```python
data_temp_new.head()
```

```python
# Рассчитаем в секундах продолжительность между первым и последним замерами.

data_temp_new['total_time'] = (data_temp_new.end_time - data_temp_new.start_time).dt.seconds
```

```python
# Удалим значения первого и последнего замеров времени, на основании которых произведены расчеты.

data_temp_new = data_temp_new.drop(columns=data_temp_new.columns[:2], axis=1)
```

<div class="alert alert-info"><b>Комментарий студента:</b>

На данном этапе на основании имеющихся сведений:
- в датасете 'data_arc' создан новый признак полной работы 'full_work'
- в датасете 'data_temp_new' удалены данные по партиям, которые имеют только 1 значение температуры, а значения температур и времени замера сагрегированы по партиям (key) по принципу "первый-последний"
- в датасете 'data_temp_new' создан новый признак полной работы 'total_time' со значениями продолжительности между первым и последним замерами температуры
- в датасетах 'data_arc' и 'data_temp_new' удалены как ненужные признаки, на основе которых производились расчеты

Поскольку при обучении моделей я буду использовать кросс-валидацию, то в целях исключения утечки целевого признака при обучении масштабирование данных я буду применять через Pipeline.

#### LinearRegression.

```python
%%time
# Применим масштабирование данных при обучении модели LinearRegression на кроссвалидации.

lr = make_pipeline(StandardScaler(), LinearRegression())

nmae = cross_val_score(lr,
                       features_train,
                       target_train,
                       scoring='neg_mean_absolute_error',
                       cv=5
                       )                               

print('Среднее значение MAE модели LinearRegression на кросс-валидации:', nmae.mean()*(-1))
```

#### RandomForestRegressor.

```python
# Случайный лес, будучи моделью на основе дерева, не чувствителен к масштабу входных объектов, поэтому масштабирование не применяем.

parameters = {'n_estimators': range(5, 21, 5),
              'max_depth': range(1, 20, 3), 
              'min_samples_leaf': range(1, 5, 2), 
              'min_samples_split': range(2, 8, 2)
             }

model_forest = RandomForestRegressor(random_state=RANDOM_STATE)

model_forest_cv = GridSearchCV(model_forest,
                               parameters,                                   
                               scoring='neg_mean_absolute_error',
                               cv=5
                              )
```

```python
%%time

# Обучение модели. 
model_forest_cv.fit(features_train, target_train)
print('Значение метрики MAE модели RandomForestRegressor:', -model_forest_cv.best_score_)
```

#### CatBoostRegressor.

```python
# Применим масштабирование данных при обучении модели CatBoostRegressor на кроссвалидации.

pipe = Pipeline([
    ('scale', StandardScaler()),    
    ('catboost', CatBoostRegressor(loss_function='MAE', silent=True, random_state=RANDOM_STATE))
])

cbr_params = {'catboost__depth': range(1, 12, 2),
             }

grid_cbr = GridSearchCV(pipe,
                        cbr_params,                            
                        scoring='neg_mean_absolute_error'
                       )
```

```python
%%time
# Обучение модели.
grid_cbr.fit(features_train, target_train)
print('Значение метрики MAE модели CatBoostRegressor:', -grid_cbr.best_score_)
```

#### LGBMRegressor.

```python
# Применим масштабирование данных при обучении модели LGBMRegressor на кроссвалидации.

pipe = Pipeline([
    ('scale', StandardScaler()),    
    ('lgbmreg', LGBMRegressor(verbose=-1,  random_state=RANDOM_STATE))
])

lgbmr_params = {'lgbmreg__n_estimators': range(50, 401, 50),
                'lgbmreg__max_depth': range(1, 10, 2),
               }

grid_lgbmr = GridSearchCV(pipe,
                          lgbmr_params,                             
                          scoring='neg_mean_absolute_error'                          
                         )
```

```python
%%time
# Обучение модели
grid_lgbmr.fit(features_train, target_train)
print('Значение метрики MAE модели LGBMRegressor:', -grid_lgbmr.best_score_)
```
